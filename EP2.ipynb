## Implementação do Método de Bisseção

```python
def bissecao(f,a,b,tol, intera, cont=0, pm=[]):
    """
     função bissecao(f, a, b, tol) , que:   
    • Recebe:
    – Uma função f ;
    – Os limites a e b do intervalo de busca inicial;
    – A tolerância para o erro tol .
    • Devolve uma aproximação para uma raiz de f , com erro menor ou igual a tol .
    """
    if f(a)*f(b) > 0.0:
        print("f(a) e f(b) tem sinais iguais, logo não é válido!")
        return None
    
    if f(a)*f(b)< 0.0:
        pm += [(a+b)/2]
     
        if cont<intera:
            if (b-a)/2 <= tol:
                pm += [(a+b)/2]
             
                if f(pm) == 0.0:
                    return pm
             
                if  f(pm)*f(a) > 0.0:
                    a = pm
                
                if f(pm)*f(b) > 0.0 :
                    b = pm
                bissecao (f,a,b,tol,intera, cont+1, pm[:])
        else:
            return pm
        return pm
```
